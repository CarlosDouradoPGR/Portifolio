## 2 - Data Cleaning (Limpeza de Dados)

```python
import pandas as pd
```

```python
df = pd.read_csv("dados.csv")
```

```python
df.head(5)
```

Verificando dados vazios:

```python
df.isna().any()
```

Nesse momento o pandas não conta os valores nulos desse dataframe, pois os valores nulos são "na" e o pandas não reconhece como valor nulo. <br>
Por isso iremos tranformar os valores "na" em ``NaN`` utilizando o replace e o numpy

```python
import numpy as np

df = df.replace("na",np.nan)
```

Verificando novamente os valores nulos

```python
df.isna().any()
```

Vamos verificar a quantidade de dados nulos em cada coluna de ``escolaridade``, ``estado_civil``, ``salario_anual`` e retirar a porcetagem de dados não informados

```python
for n in ["escolaridade", "estado_civil", "salario_anual"]:
    print(f'Quantidade de valores nulos em {n}: {df[n].isnull().sum()}, { round(df[n].isnull().sum() * 100 / len(df[n]), 2)}% ')
```

Agora veremos o quanto a queda dos clientes que tem valores nulos ira after na analise final dos dados.

Busacando dados sobre os tipos de cada coluna

Com as informações dos dados do dataframe, precisamos tratar algumas colunas:
- ``limite_credito`` 
- ``valor_transacoes_12m``

```python
df[["limite_credito", "valor_transacoes_12m"]]
```

Criando uma função ``lambda`` para mudar o tipo da coluna

```python
trade_type = lambda x: float(x.replace(".", "").replace(",", "."))
```

Testando:

```python
df["limite_credito"].apply(trade_type)
```

Realizando as mundaças

```python
df["limite_credito"] = df["limite_credito"].apply(trade_type)
df["valor_transacoes_12m"] = df["valor_transacoes_12m"].apply(trade_type)
```

## 4 - Vizualização

Para facilitar essa vizualização, iremos utilizar a frequência absoluta, para entender mais sobre a estratégia consulte: https://www.todamateria.com.br/frequencia-absoluta/

Agora vamos importar as bibliotecas de vizualicação e criar os DataFrames inadimplantes e adimplates

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
df_adimplente = df[df["default"] == 0]
```

```python
df_inadimplente = df[df["default"] == 1]
```

```python
df_adimplente.head(4)
```

```python
df_inadimplente.head(4)
```

### 4.1 Dados categóricos

Nesse ponto iremos vizualizar a relacão entre adimplentes e inadimplentes e as colunas ``["escolaridade, "salário_anual"]``, que são atributos categóricos

- Escolaridade

```python
df.select_dtypes("object").describe()
```

```python
coluna = "escolaridade"
titulos = ["Escolaridade dos Clientes", "Escolaridade dos Clientes Adimplentes", 'Escolaridade dos clientes Inadimplentes']

eixo = 0
max_y = 0
max = df.select_dtypes("object").describe()[coluna]['freq']

fig, ax = plt.subplots(1, 3, figsize=(20,5), sharex = True)

for dataframe in [df, df_adimplente, df_inadimplente]:
    
    df_to_plot = dataframe[coluna].value_counts().to_frame()
    df_to_plot.rename(columns = {coluna: 'frequencia_absoluta'}, inplace=True)
    df_to_plot[coluna] = df_to_plot.index
    df_to_plot.sort_values(by=[coluna], inplace=True)
    df_to_plot.sort_values(by=[coluna])
    
    f = sns.barplot(x=df_to_plot[coluna], y=df_to_plot['frequencia_absoluta'], ax=ax[eixo])
    f.set(title=titulos[eixo], xlabel=coluna.capitalize(), ylabel = "frequencia_absoluta")
    f.set_xticklabels(labels=f.get_xticklabels(), rotation=90)
    
    _, max_y_f = f.get_ylim()
    max_y = max_y_f if max_y_f > max_y else max_y
    f.set(ylim=(0, max_y))
    
    eixo += 1
```

- Salário Anual

```python
coluna = 'salario_anual'
titulos = ['Salário Anual dos Clientes', 'Salário Anual dos Clientes Adimplentes', 'Salário Anual dos Clientes Inadimplentes']

eixo = 0
max_y = 0
figura, eixos = plt.subplots(1,3, figsize=(20, 5), sharex=True)

for dataframe in [df, df_adimplente, df_inadimplente]:

  df_to_plot = dataframe[coluna].value_counts().to_frame()
  df_to_plot.rename(columns={coluna: 'frequencia_absoluta'}, inplace=True)
  df_to_plot[coluna] = df_to_plot.index
  df_to_plot.reset_index(inplace=True, drop=True)
  df_to_plot.sort_values(by=[coluna], inplace=True)

  f = sns.barplot(x=df_to_plot[coluna], y=df_to_plot['frequencia_absoluta'], ax=eixos[eixo])
  f.set(title=titulos[eixo], xlabel=coluna.capitalize(), ylabel='Frequência Absoluta')
  f.set_xticklabels(labels=f.get_xticklabels(), rotation=90)
  _, max_y_f = f.get_ylim()
  max_y = max_y_f if max_y_f > max_y else max_y
  f.set(ylim=(0, max_y))
  eixo += 1
```

### 4.2 Dados Númericos

Nesse ponto iremos vizualizar a relação entre os cliente adimplentes e inadimplentes e os atribtos númericos 

```python
df.drop(['id', 'default'], axis=1).select_dtypes('number').head(n=5)
```

- Quantidade de transações nos útimos 12 meses

```python
coluna = 'qtd_transacoes_12m'
titulos = ['Qtd. de Transações no Último Ano', 'Qtd. de Transações no Último Ano de Adimplentes', 'Qtd. de Transações no Último Ano de Inadimplentes']

eixo = 0
max_y = 0
figura, eixos = plt.subplots(1,3, figsize=(20, 5), sharex=True)

for dataframe in [df, df_adimplente, df_inadimplente]:

  f = sns.histplot(x=coluna, data=dataframe, stat='count', ax=eixos[eixo])
  f.set(title=titulos[eixo], xlabel=coluna.capitalize(), ylabel='Frequência Absoluta')

  _, max_y_f = f.get_ylim()
  max_y = max_y_f if max_y_f > max_y else max_y
  f.set(ylim=(0, max_y))

  eixo += 1
```

- Valor das Transações nos Últimos 12 Meses

```python
coluna = 'valor_transacoes_12m'
titulos = ['Valor das Transações no Último Ano', 'Valor das Transações no Último Ano de Adimplentes', 'Valor das Transações no Último Ano de Inadimplentes']

eixo = 0
max_y = 0
figura, eixos = plt.subplots(1,3, figsize=(20, 5), sharex=True)


for dataframe in [df, df_adimplente, df_inadimplente]:

  f = sns.histplot(x=coluna, data=dataframe, stat='count', ax=eixos[eixo], color="orange")
  f.set(title=titulos[eixo], xlabel=coluna.capitalize(), ylabel='Frequência Absoluta')

  _, max_y_f = f.get_ylim()
  max_y = max_y_f if max_y_f > max_y else max_y
  f.set(ylim=(0, max_y))

  eixo += 1
```

 - Valor de Transações nos Últimos 12 Meses x Quantidade de Transações nos Últimos 12 Meses

```python
f = sns.relplot(x='valor_transacoes_12m', y='qtd_transacoes_12m', data=df, hue="default")
_ = f.set(
    title='Relação entre Valor e Quantidade de Transações no Último Ano', 
    xlabel='Valor das Transações no Último Ano', 
    ylabel='Quantidade das Transações no Último Ano'
)
```
