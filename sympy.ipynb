```python
import sympy as sp

# Definindo as variáveis simbólicas
i, w, mu, k_x, k_y, sigma_z, sigma_x, sigma_y = sp.symbols('i w mu k_x k_y sigma_z sigma_x sigma_y')

# Definindo as matrizes A1 e A2
A1 = sp.Matrix([[i*w*mu - (k_x**2)/sigma_z, -k_x*k_y/sigma_z],
                [-k_x*k_y/sigma_z, i*w*mu - k_y**2/sigma_z]])

A2 = sp.Matrix([[-sigma_x - i*k_y**2/(w*mu), i*k_x*k_y/(w*mu)],
                [i*k_x*k_y/(w*mu), -sigma_y - i*k_x**2/(w*mu)]])

# Definindo a matriz A
A = sp.Matrix([[sp.zeros(2), A1], [A2, sp.zeros(2)]])

# Calculando os autovalores e autovetores
eigenvals = A.eigenvals()
eigenvects = A.eigenvects()

# Exibindo os resultados
print(eigenvals)

print(eigenvects)
```
